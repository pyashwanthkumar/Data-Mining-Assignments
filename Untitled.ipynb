```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
%matplotlib inline
from scipy.spatial.distance import cdist
from math import *
from sklearn import preprocessing
```

```python
def cos_sim(a,b):
    dot_prod = np.dot(a,b)
    a_norm = np.linalg.norm(a)
    b_norm = np.linalg.norm(b)
    return dot_prod/(a_norm * b_norm)
def manhattan_distance(x,y):
    return sum(abs(a-b) for (a,b) in (x,y))
```

```python
df = pd.read_csv("income_tr.csv")
df_scale = preprocessing.scale(df['fnlwgt'])
        

c =[]
d =[]
for i in range(0,len(df)):
    for j in range(0,len(df)-1):
        a = df_scale[i]
        b = df_scale[j]
        if i!=j and (b-a)!=0:
            c.append(abs(b-a))
    d.append(min(c))
print(d)
        
```
